```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
month_steps_med = fitbit.groupby('Months_encoded')['Steps'].median().tolist()
months = list(range(12))

fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(12,4))
ax[0].barh(months, month_steps_med);

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
weekday_steps_med = fitbit.groupby('Work_or_Weekend')['Steps'].median()
weekend_or_work = [0,1]

ax[1].barh(weekend_or_work, weekday_steps_med);
```
